```python
# Fields Medal Data Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read data from CSV file
df = pd.read_csv('fieldMedalists.csv', encoding = 'cp1252', index_col = 1)

totalWinners = df['Gender'].count()
print(totalWinners)

# Total number of winners by gender
WinnersByGender = pd.pivot_table(data = df, index = ['Gender'], columns = ['Gender'], aggfunc = 'count')
WinnersByGender = df.groupby('Gender')['Gender'].value_counts()
# print(WinnersByGender)
WinnersByGender.plot(kind = 'bar') # Call Matplotlib in backend

# Total number of winners by country
WinnersByCountry = pd.pivot_table(data = df, index = ['Citizenship'], aggfunc = 'count')
WinnersByCountry.plot(kind = 'bar', legend = None)

# Total number of winners in each year
WinnersByYear = pd.pivot_table(data = df, index = ['Year'], aggfunc = 'count')
WinnersByYear.plot(kind = 'line')

# Winners by age
df['Age'] = df['Year'] - df['Born']
WinnersByAge = pd.pivot_table(data = df, index = ['Year'], values = 'Age')
WinnersByAge.plot(kind = 'line')
```
